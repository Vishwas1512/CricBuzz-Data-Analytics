# Web Scraping from CricBuzz(ICC MENS CRICKET2024)

**Import necessary libraries**

```python
import requests
from bs4 import BeautifulSoup
```

**Get Data from Website**

```python
r=requests.get('https://www.cricbuzz.com/cricket-series/7476/icc-mens-t20-world-cup-2024/matches')
r
```

**Verify the content from website**

```python
r.content
```

**Create a Soup**

```python
soup=BeautifulSoup(r.content, 'html.parser')
soup
```

**Extracting Match Name**

```python
Match_name=[i.text for i in soup.select(' div.cb-col-75.cb-col > div.cb-col-60.cb-col.cb-srs-mtchs-tm > a.text-hvr-underline > span')]
```

```python
Match_name
```

**Extracting Match Venue**

```python
Match_Venue=[i.text for i in soup.select('div.cb-col-75.cb-col > div.cb-col-60.cb-col.cb-srs-mtchs-tm > div')]
```

**Extracting Match Results**

```python
Match_Results=[i.text for i in soup.select('div.cb-col-75.cb-col > div.cb-col-60.cb-col.cb-srs-mtchs-tm > a.cb-text-complete')]
```
